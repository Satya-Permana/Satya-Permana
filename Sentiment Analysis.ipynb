```python
def cleaning(text):
    # Menghapus mention (contoh: @username)
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)

    # Menghapus hashtag (contoh: #hashtag)
    text = re.sub(r'#\w+', '', text)

    # Menghapus retweet marker (contoh: RT)
    text = re.sub(r'RT[\s]+', '', text)

    # Menghapus URL
    text = re.sub(r'https?://\S+', '', text)

    # Menghapus karakter selain huruf, angka, dan spasi
    text = re.sub(r'[^A-Za-z0-9\s]', '', text)

    # Menghapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    # Menghapus tanda tanya berlebih
    text = re.sub(r'\?+', '?', text)

    # Menghapus titik berlebih
    text = re.sub(r'\.+\.', '.', text)

    # Menghapus koma berlebih
    text = re.sub(r',+,', ',', text)

    return str(text)
```

```python
k=k.apply(cleaning)
```

```python
k=k.str.lower()
```

```python
more_stopwords = ["tidak"]

stop_words = StopWordRemoverFactory().get_stop_words()
stop_words.extend(more_stopwords)

new_array = ArrayDictionary(stop_words)
stop_words_remover_new = StopWordRemover(new_array)

def stopword(str_text):
    str_text = stop_words_remover_new.remove(str_text)
    return str_text

k = k.apply(lambda x: stopword(x))

k
```

```python
tokenizing = k.apply(lambda x: x.split())
tokenizing
```

```python
def stemming(text_cleaning):
    factory = StemmerFactory()

    stemmer = factory.create_stemmer()

    do = []

    for w in text_cleaning:
        dt = stemmer.stem(w)
        do.append(dt)

    d_clean = []

    d_clean = " ".join(do)

    print(d_clean)

    return d_clean

tokenizing = tokenizing.apply(stemming)
tokenizing.to_csv('x2.csv')
```

```python
def convert(text):
    translator = Translator(to_lang="en", from_lang="id")
    translation = translator.translate(text)
    return translation

data['eng'] = data['Tweet Text'].apply(convert)

data.to_csv('x3.csv')
```
